```python
#Layer-wise하게 !!!

from sklearn import preprocessing

st_scaler = preprocessing.StandardScaler()

A_W0_2d = A_W0.reshape(-1, 1)
A_W1_2d = A_W1.reshape(-1, 1)
A_W2_2d = A_W2.reshape(-1, 1)
A_W3_2d = A_W3.reshape(-1, 1)
A_W4_2d = A_W4.reshape(-1, 1)

new_A_W0 = st_scaler.fit_transform(A_W0_2d)
new_A_W1 = st_scaler.fit_transform(A_W1_2d)
new_A_W2 = st_scaler.fit_transform(A_W2_2d)
new_A_W3 = st_scaler.fit_transform(A_W3_2d)
new_A_W4 = st_scaler.fit_transform(A_W4_2d)

scaled_total_A_W = []
scaled_total_A_W = (new_A_W0 + new_A_W1 + new_A_W2 + new_A_W3 + new_A_W4)/5
scaled_yerr_total_A_W = []


for i in range(50):
    scaled_yerr_total_A_W.append(np.std([new_A_W0, new_A_W1, new_A_W2, new_A_W3, new_A_W4]))

plt.figure(figsize=(10,8))
#plt.title("Weight Variance", fontsize=18)
plt.xlabel('Global Epoch', fontsize=19)
plt.ylabel('Variance of accumulated Weight', fontsize=19)
#plt.xlim(1,18)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.plot(x1, scaled_total_A_W, marker='o', c='r', linewidth=4, label='Weights of average layers')
plt.errorbar(x1, scaled_total_A_W, yerr=scaled_yerr_total_A_W, label='Deviation of weights', marker='o', c='black')

plt.legend(fontsize=18)
plt.grid(True)
plt.show()
```

# Acc, Loss

```python
x = np.arange(0,51)

for i in range(tests):
    globals()['for_acc_try{}'.format(i)] = []
    globals()['for_loss_try{}'.format(i)] = []
    
    for g_epoch in range(51):
        globals()['for_acc_try{}'.format(i)].append(np.mean(globals()['t{}_ACC'.format(i)][g_epoch]))
        globals()['for_loss_try{}'.format(i)].append(np.mean(globals()['t{}_LOSS'.format(i)][g_epoch]))
        
final_acc, final_loss = 0, 0

for i in range(tests):
    globals()['for_acc_try{}'.format(i)] = np.array(globals()['for_acc_try{}'.format(i)])
    globals()['for_loss_try{}'.format(i)] = np.array(globals()['for_loss_try{}'.format(i)])
    
    final_acc = final_acc + globals()['for_acc_try{}'.format(i)]
    final_loss = final_loss + globals()['for_loss_try{}'.format(i)]

final_acc, final_loss = final_acc/tests, final_loss/tests

plt.figure(figsize=(10,8))
#plt.title("Accuracy", fontsize=18)
plt.xlabel('Global Epoch', fontsize=19)
plt.ylabel('Test Accuracy', fontsize=19)
plt.xlim(0,50)
plt.ylim(0,1)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)

plt.plot(x, final_acc, marker='o', c='r', linewidth=2, label='Acc')
plt.plot(x, final_loss, marker='o', c='b', linewidth=2, label='Loss')

plt.legend(fontsize=20)
plt.grid(True)
plt.show()
```
